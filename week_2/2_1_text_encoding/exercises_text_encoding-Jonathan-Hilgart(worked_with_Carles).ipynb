2) ELT (Extract, Transform, Load): A text file of emojis into a Python dict
---

A) Extract the emoji table from coropora folder (aka, read from the file `emoji_table.txt`)

```python
with open('../../corpora/emoji_table.txt','r') as f:
    emoji = f.read()
emoji
```

B) Transform the data (aka, mungle)

`'👪,E_family'` is okay  

`'👨\u200d👩\u200d👧,E_family1'` is __not__ okay. It should be `'👨👩👧👦,E_family2'`

```python
import re
```

```python
#Two things to split on, first the /n, then create a list of lists by splitting on ,
emoji_fixed = emoji.replace("\u200d",'')
emoji_fixed = emoji_fixed.lower()
emoji_fixed  = emoji_fixed.split('\n')
emoji_fixed.remove(emoji_fixed[0])

emoji_final = []
for line in emoji_fixed:
    if line == 'emoji':
        pass
    elif line =='meaning':
        pass
    else:
        emoji_final.append(line.split(','))
    
emoji_final[-10:]
```

C) Load into a Python dictionary where the keys are text labels and the values are emojis.

```python
#create a dict by swapping the order of our list of lists
emoji_map = {}

for k in emoji_final:
    emoji_map[k[1]]=k[0]
emoji_map
```

```python
# Tests for your final data structure
assert emoji_map['d_ok'] == '👌'
assert emoji_map['d_poop'] == '💩'
```
